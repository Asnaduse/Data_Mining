```python
import pandas as pd
data_set=pd.read_excel("financial_sample.xlsx")
data_set
```

```python
data_set.rename(columns={"product":"product_name","country":"city","Discount Band":"category","Units Sold":"price"}, inplace = True)
```

```python
data_set.rename(columns = {"Sale Price":"quantity", "Month Name":"month", "Date":"date", "Year":"year", "Segment":"customer_name"}, inplace = True)
```

```python
data_set.rename(columns = {"Manufacturing Price":"manufacturing_price"}, inplace = True)
```

```python
data_set.head(3)
```

```python
data_set.columns=data_set.columns.str.strip().str.lower().str.replace('NaN','low')
data_set
```

```python
data_set.isnull().sum()
```

```python
cleaned_data=data_set.copy()
```

```python
cleaned_data.loc[:, 'category'] = cleaned_data['category'].fillna('low')
```

```python
cleaned_data.isnull().sum()
```

```python
product = cleaned_data[['product', 'category']]
product.to_csv("product_data.csv", index=False)
productnew = pd.read_csv("product_data.csv")
productnew.head()
```

```python
customer = cleaned_data[['customer_name','country']]
customer.to_csv("customer_data.csv", index=False)
customernew = pd.read_csv("customer_data.csv")
customernew.head()
```

```python
date=cleaned_data[['date','month','year']]
date.to_csv("date_data.csv", index=False)
datenew=pd.read_csv("date_data.csv")
datenew.head()
```

```python
sales=cleaned_data[['quantity','price']]
sales.to_csv("sales_data.csv")
salesnew=pd.read_csv("sales_data.csv")
salesnew.head()
```
